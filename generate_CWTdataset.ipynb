```python
import pickle
import pywt
import matplotlib.pyplot as plt
import numpy as np
# from signal_data_add import *
```

```python
import pickle as pk
f = open('RML2016.10a_dict.pkl','rb')
data = pk.load(f, encoding = 'latin1')
```

```python
CWTdata = dict()
for i in data.keys():
    pp = []
    for j in range(len(data[i])):
        coefs1, freqs = pywt.cwt(data[i][j][0], np.arange(1, 100), 'morl')
        coefs2, freqs = pywt.cwt(data[i][j][1], np.arange(1, 100), 'morl')
        coef = np.expand_dims(np.stack([coefs1,coefs2]),0)
        pp.append(coef)
    print(i,"finished")
    pp = np.vstack(pp)
    CWTdata[i] = pp
        

with open('CWTdata.pkl', 'wb') as f:
    pickle.dump(CWTdata, f)


```
